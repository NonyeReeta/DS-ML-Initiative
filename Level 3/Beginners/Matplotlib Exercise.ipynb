```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

company_sales = pd.read_csv("company_sales_data.csv")
font_dict = {
    "family": "Times New Roman",
    "color": "black",
    "weight": "bold",
    "size": 18
}

"""Exercise 1: Read total profit of all months and show it using a line plot"""
total_sales = company_sales.total_profit
month_num = company_sales.month_number
plt.plot(month_num, total_sales)
plt.xticks(month_num)
plt.title("Company Profit per Month", fontdict=font_dict)
plt.xlabel("Month Number", fontdict=font_dict)
plt.ylabel("Total Profit", fontdict=font_dict)
plt.show()
```

```python
"""Exercise 2: Read all product sales data and show it using multi line plot"""
facecream_sales = company_sales.facecream
facewash_sales = company_sales.facewash
toothpaste_sales = company_sales.toothpaste
bathingsoap_sales = company_sales.bathingsoap
shampoo_sales = company_sales.shampoo
moisturizer_sales = company_sales.moisturizer

plt.plot(month_num, facecream_sales, "o-", label="Face Cream Sales Data", linewidth=4)
plt.plot(month_num, facewash_sales, "o-", label="Face Wash Sales Data", linewidth=4)
plt.plot(month_num, toothpaste_sales, "o-", label="Toothpaste Sales Data", linewidth=4)
plt.plot(month_num, bathingsoap_sales, "o-", label="Bathing Soap Sales Data", linewidth=4)
plt.plot(month_num, shampoo_sales, "o-", label="Shampoo Sales Data", linewidth=4)
plt.plot(month_num, moisturizer_sales, "o-", label="Moisturizer Sales Data", linewidth=4)
plt.xticks(month_num)
plt.title("Sales Data", fontdict=font_dict)
plt.ylabel("Sales Unit in numbers", fontdict=font_dict)
plt.xlabel("Month Number", fontdict=font_dict)
plt.legend(loc="best", title="Products", bbox_to_anchor=(1.6, 1.0))
plt.show()
```

```python
"""Question 3: read toothpaste sales data of each month and show it using a scatter plot"""
plt.scatter(month_num, toothpaste_sales, label="Toothpaste Sales Data", color="b")
plt.xticks(month_num)
plt.title("Toothpaste Sales Data", fontdict=font_dict)
plt.xlabel("Month Number", fontdict=font_dict)
plt.ylabel("Number of units sold", fontdict=font_dict)
plt.grid(color="k", linestyle="--")
plt.legend()
plt.show()
```

```python
"""Question 4: Read face cream and facewash product sales data and show it using a bar chart"""
x_axis = np.arange(len(month_num))
plt.bar(x_axis - 0.2, facecream_sales, label="Face Cream sales data", width=0.4, color="b")
plt.bar(x_axis + 0.2, facewash_sales, label="Face Wash sales data", width=0.4, color="y")
plt.xticks(x_axis, month_num)
plt.xlabel("Month Number", fontdict=font_dict)
plt.ylabel("Sales unit in number", fontdict=font_dict)
plt.title("Facewash and Facecream Sales Data", fontdict=font_dict)
plt.grid(color="k", linestyle="--")
plt.legend()
plt.show()
```

```python
"""Question 5: Calculate total sales data for last year for each product in percentage"""
total = company_sales[["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]].agg(sum)
labels = ["Facecream", "Facewash", "Toothpaste", "Bathingsoap", "Shampoo", "Moisturizer"]
plt.figure(figsize=(8, 8))
plt.pie(total, labels=labels, startangle=90, autopct="%0.1f%%", textprops={"fontsize": 10})
plt.title("Sales Data", fontdict=font_dict)
plt.legend(loc=4, title="Products", bbox_to_anchor=(1.3, 0.1))
plt.show()
```
